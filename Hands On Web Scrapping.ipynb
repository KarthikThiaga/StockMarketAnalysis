```python
from bs4 import BeautifulSoup
import pandas as pd
import requests
from urllib.parse import urljoin, urlparse, parse_qs, urlencode
```

```python
## Function to scrape data

def scrape_data(url):
    Final_data = []
    current_page = 2
    while url:
        print(url)
        response = requests.get(url)
        html_content = response.text
        
        soup = BeautifulSoup(html_content, 'html.parser')
        data_on_page = extract_data(soup)
        Final_data.extend(data_on_page)
        next_page_link = find_next_page_link(soup,url,current_page)
        url = next_page_link if next_page_link else None
       
    # Create data frame 
    
        Final_Columns = ['S.No.',
                         'Name',
                         'CMP Rs.',
                         'P/E',
                         'Mar Cap Rs.Cr.',
                         'Div Yld %',
                         'NP Qtr Rs.Cr.',
                         'Qtr Profit Var %',
                         'Sales Qtr Rs.Cr.',
                         'Qtr Sales Var %',
                         'ROCE %']
        Final_df = pd.DataFrame(Final_data,columns=Final_Columns)
    return Final_df

def extract_data(soup):
    data_list = []
    table = soup.find('table', class_='data-table text-nowrap striped mark-visited')
    for row in table.find_all('tr'):
        row_data = [cell.text.strip() for cell in row.find_all('td')]
        data_list.append(row_data)
    
    return data_list

def find_next_page_link(soup, base_url, current_page):
    # Find any link within <a> tags with an href attribute containing "?page="
    next_page_link = soup.find('a', href=lambda href: href and '?page=' in href)
    
    if next_page_link:
        next_page_url = next_page_link['href']
        
        # Extract the current page number
        parsed_url = urlparse(next_page_url)
        query_params = parse_qs(parsed_url.query)
        current_page = int(query_params.get('page', [current_page])[0])

        # Increment the page number
        next_page_number = current_page + 1

        # Update the query parameters with the new page number
        query_params['page'] = [next_page_number]

        # Reconstruct the URL with the updated query parameters
        absolute_url = urljoin(base_url, urlunparse(parsed_url._replace(query=urlencode(query_params, doseq=True))))

        print("Next Page URL:", absolute_url)
        return absolute_url

    return None
    
```

```python
url_to_scrape = 'https://www.screener.in/company/CNX500/'
Nifty_500_df = scrape_data(url_to_scrape)
```

```python
import re
from urllib.parse import urlparse, urlunparse, parse_qs, urlencode

def find_next_page_link(soup, base_url):
    # Find the next page link within <a> tags
    next_page_link = soup.find('a', href=re.compile(r'\?page=\d+'))
    if next_page_link:
        next_page_url = next_page_link['href']
        
        # Extract the page number from the href using a regular expression
        page_number_match = re.search(r'\?page=(\d+)', next_page_url)
        if page_number_match:
            page_number = page_number_match.group(1)
            
            # Parse the base URL
            parsed_url = urlparse(base_url)
            
            # Update the query parameters with the new page number
            query_params = parse_qs(parsed_url.query)
            query_params['page'] = [page_number]
            
            # Reconstruct the URL with the updated query parameters
            updated_url = urlunparse(parsed_url._replace(query=urlencode(query_params, doseq=True)))

            print("Next Page URL:", updated_url)
            return updated_url

    return None
```
